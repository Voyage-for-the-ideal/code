```python
#快速傅里叶变换
import math
import cmath

def f(t):
    ans=[]
    n=len(t)
    for i in range(n):
        ans.append(0.7*math.sin(4*math.pi*t[i])+math.sin(10*math.pi*t[i]))
    return ans

def fft(x):
    #x 输入的序列
    n=len(x)
    g=[0]*n
    if n==1:
        return x
    wn=cmath.exp(-2j*math.pi/n)
    w=1
    even = (x[0::2])
    odd = (x[1::2])
    g0=fft(even)
    g1=fft(odd)
    for k in range(n//2):
        g[k]=(g0[k]+w*g1[k])/2
        g[k+n//2]=(g0[k]-w*g1[k])/2
        w=w*wn
    return g

values=[128,256]
for n in values:
    t=[]
    for i in range(n):
        t.append(i/n)
    # 计算函数值
    x = f(t)
    # 执行快速傅里叶变换
    g = fft(x)
    # 输出结果
    for i in range(len(g)):
        print(f"向量g的第{i+1}个分量, x_{i+1}={t[i]}, y_{i+1}={g[i]:.10f}")
    print()

```

```python
#傅里叶变换
def fourier(x,n):
    g=[]
    temp=0
    for i in range(n):
        for k in range(n):
            temp+=x[i]*cmath.exp(-2j*math.pi*i*k/n)
        temp=temp/n
        g.append(temp)
    return g

values=[128,256]
for n in values:
    t=[]
    for i in range(n):
        t.append(i/n)
    # 计算函数值
    x = f(t)
    g=fourier(x,n)
    for i in range(n):
        print(f"{g[i]}")
```
