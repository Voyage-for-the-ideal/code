```python
def main():
    import sys
    N=int(input())
    input_ = sys.stdin.read().split()
    ptr=0
    temp=0
    found_error=0
    m=int(len(input_)/(3*N**2))
    for i in range(m):
        for x in range(N):
            if found_error:
                break
            for y in range(N):
                if found_error:
                    break
                for j in range(N):
                    temp+=int(input_[i*3*N**2+x*N+j])*int(input_[(i*3+1)*N**2+j*N+y])
                if temp!=int(input_[(i*3+2)*N**2+x*N+y]):    
                    found_error=1
                    break
                temp=0
        print("Yes" if not found_error else "No")
        found_error=0

if __name__ == "__main__":
    main()
```

```python
import sys

def main():
    data = list(map(int, sys.stdin.read().split()))
    ptr = 0
    N = data[ptr]
    ptr += 1
    while ptr < len(data):
        # 读取矩阵A
        A = []
        for _ in range(N):
            row = data[ptr:ptr+N]
            ptr += N
            A.append(row)
      
        # 读取矩阵B
        B = []
        for _ in range(N):
            row = data[ptr:ptr+N]
            ptr += N
            B.append(row)
      
        # 读取矩阵C
        C = []
        for _ in range(N):
            row = data[ptr:ptr+N]
            ptr += N
            C.append(row)
      
        result = "Yes"
        for i in range(N**2):
            x=i%N
            y=i//N
            ans=0
            for j in range(N):
                ans+=A[x][j]*B[j][y]
            if ans != C[x][y]:
                result = "No"
                break  # 只要有一次失败，即可判定为No
      
        print(result)

if __name__ == "__main__":
    main()
```
